#### Frequency Domain

```python
import numpy as np 
import matplotlib.pyplot as plt 

from scipy import pi 
from scipy.fftpack import fft

import librosa
import librosa.display
```

```python
audio = 'bye'
data, sampling_rate = librosa.load('./{}.wav'.format(audio))
```

```python
signalAmplitude = np.sin(data)
signalAmplitude
```

```python
plt.subplot(211)
plt.plot(data, signalAmplitude, 'bs')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.subplot(212)
plt.magnitude_spectrum(signalAmplitude, Fs=4)
plt.show()
```
